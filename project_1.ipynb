```python


# Assuming churned_subscribers is your DataFrame containing churned data

# Churn Rate (you need the total number of subscribers for this)
total_subscribers = df.shape[0]  # Replace with the actual total number of subscribers if known
churn_rate = (churned_data.shape[0] / total_subscribers) * 100
print(f"Churn Rate: {churn_rate:.2f}%")

# Churn Reasons Distribution
churn_reasons_distribution = churned_data['Reason_for_Churn'].value_counts(normalize=True) * 100
print("Churn Reasons Distribution:")
print(churn_reasons_distribution)

# Subscription Type Churn Rate
subscription_type_churn_rate = churned_data['Subscription_Type'].value_counts(normalize=True) * 100
print("Subscription Type Churn Rate:")
print(subscription_type_churn_rate)

# Average Subscription Duration
average_subscription_duration = churned_data['Subscription_Duration'].mean()
print(f"Average Subscription Duration: {average_subscription_duration:.2f} days")

# Churn by Age Group
bins = [18, 30, 40, 50, 60, 70]
labels = ['18-30', '30-40', '40-50', '50-60', '60-70']
churned_data['Age_Group'] = pd.cut(churned_data['Age'], bins=bins, labels=labels, right=False)
churn_by_age_group = churned_data['Age_Group'].value_counts(normalize=True) * 100
print("Churn by Age Group:")
print(churn_by_age_group)

# Average Monthly Hours Watched
average_monthly_hours_watched = churned_data['Monthly_Hours_Watched'].mean()
print(f"Average Monthly Hours Watched: {average_monthly_hours_watched:.2f} hours")

# Customer Lifetime Value Loss (assuming you have monthly payment and duration data)
cltv_loss = churned_data['Monthly_Payment'].mean() * churned_data['Subscription_Duration'].mean()
print(f"Average Customer Lifetime Value Loss: ${cltv_loss:.2f}")
```

```python


# Create a dictionary for age groups
age_groups = {
    '18-30': (18, 30),
    '30-40': (30, 40),
    '40-50': (40, 50),
    '50-60': (50, 60),
    '60-70': (60, 70)
}

# Function to map age to age group
def map_age_to_group(age):
    for group, (lower, upper) in age_groups.items():
        if lower <= age < upper:
            return group
    return None

# Assuming churned_subscribers is your DataFrame containing the churned subscribers data

# Apply the function to create the Age_Category column
churned_data['Age_Category'] = churned_data['Age'].apply(map_age_to_group)

# Group by 'Subscription_Type', 'Reason_for_Churn', and 'Age_Category'
grouped_data = churned_data.groupby(['Subscription_Type', 'Reason_for_Churn', 'Age_Category']).size().reset_index(name='Count')

# Create the bar plot using seaborn
plt.figure(figsize=(14, 8))
sns.barplot(x='Subscription_Type', y='Count', hue='Age_Category', data=grouped_data, edgecolor='black')

# Customize the plot
plt.xlabel('Subscription Type and Reason for Churn')
plt.ylabel('Number of Subscribers')
plt.title('Number of Subscribers by Subscription Type, Reason for Churn, and Age Category')
plt.legend(title='Age Category', bbox_to_anchor=(1.05, 1), loc='upper left')


plt.show()
```
